```python
df1 = pd.read_csv("../raw/google_patents_dataset.csv")  #first dataset 
df1.head() #shows dataset
```

```python
df2 = pd.read_csv("../raw/epo_patent.csv") #2.dataset
df2.head() #shows dataset
```

```python
# If 'patent_id' does not exist in df2, create it and assign NaN values
df2["patent_id"] = df2.get("patent_id", np.nan) 

# Fill missing values in the 'title' column with an empty string (handling potential nulls)
df2["title"] = df2["title"].fillna("")  

# If 'abstract' column is missing in df2, add it and fill with NaN
df2["abstract"] = df2.get("abstract", np.nan)

# If 'assignee' column is missing in df2, add it and fill with NaN
df2["assignee"] = df2.get("assignee", np.nan)

# If 'inventor_author' column is missing in df2, add it and fill with NaN
df2["inventor_author"] = df2.get("inventor_author", np.nan)

# If 'priority_date' column is missing in df2, add it and fill with NaN
df2["priority_date"] = df2.get("priority_date", np.nan)

# Ensure other date columns exist, filling with NaN if missing
df2["filing_date"] = df2.get("filing_date", np.nan)
df2["publication_date"] = df2.get("publication_date", np.nan)
df2["grant_date"] = df2.get("grant_date", np.nan)

# Ensure link columns exist, filling with NaN if missing
df2["result_link"] = df2.get("result_link", np.nan)
df2["image_link"] = df2.get("image_link", np.nan)

# Set default values for categorical columns if they are missing ('unknown' or 'other')
df2["language"] = df2.get("language", "unknown")
df2["technology_category"] = df2.get("technology_category", "other")

# Ensure year columns exist, filling with NaN if missing
df2["publication_year"] = df2.get("publication_year", np.nan)
df2["filing_year"] = df2.get("filing_year", np.nan)

# Reorder the columns of df2 to match the exact order of df1 columns
df2 = df2[df1.columns]
```

```python
# Concatenate (combine) the dataframes together
# ignore_index=True resets the index, ignoring the original indices of the source dataframes to create a new one
merged_df = pd.concat([df1, df2], ignore_index=True)

print("Merged Shape:", merged_df.shape)
merged_df.head()
```

```python
# Print the dimensions of the data (rows, columns)
print("Veri Boyutu:", merged_df.shape)

# Use 'list' to learn which columns the dataframe consists of
print("\nSütunlar:", list(merged_df.columns))
print("\nVeri Türleri:\n")
print(merged_df.dtypes)

# Missing value analysis
print("\nEksik Değer Sayıları:")
missing = merged_df.isnull().sum()
missing = missing[missing > 0]
if len(missing) > 0:
    print(missing)
else:
    print("Hiç eksik değer yok")

# Missing value percentage for each column (%)
print("\nEksik Değer Yüzdeleri:")
missing_percent = merged_df.isnull().mean() * 100
# SORTS FROM HIGHEST MISSING VALUE COUNT TO LOWEST
print(missing_percent[missing_percent > 0].sort_values(ascending=False))

# Sample rows
print("\nÖrnek Satırlar:")
display(merged_df.head(5))

# Number of unique values in each column
print("\nHer sütundaki benzersiz değer sayısı:")
print(merged_df.nunique())
# NUNIQUE = RETURNS THE COUNT OF UNIQUE VALUES.
# UNIQUE = SHOWS THE ACTUAL UNIQUE VALUES DIRECTLY.

# Language distribution
print("\nDil Dağılımı:")
print(merged_df['language'].value_counts())

# Technology category distribution
print("\nTeknoloji Kategorileri:")
print(merged_df['technology_category'].value_counts())

# Check year columns (are they float or int?)
print("\nYıl Sütunları (ilk 5):")
print(merged_df[['publication_year', 'filing_year']].head())

# Inspect 'image_link' entries that are NaN
nan_images = merged_df[merged_df['image_link'].isna()] # RETURNS TRUE OR FALSE
print(f"\nResim linki eksik satır sayısı: {len(nan_images)}") # RETURNS THE ONES THAT ARE TRUE
if len(nan_images) > 0:
    display(nan_images[['patent_id', 'title', 'language', 'technology_category']])
# DISPLAY -> SHOWS AS A TABLE
```

```python
# Calculate the total number of missing (null) values for each column in the dataframe
merged_df.isnull().sum()
```

```python
# Remove rows where critical columns like 'patent_id', 'title', or 'language' contain missing values (NaN)
merged_df = merged_df.dropna(subset=['patent_id','title','language'])
```

```python
# Calculate the total number of missing (null) values for each column in the dataframe
merged_df.isnull().sum()
```

```python
# Small text missing values -> fill with 'unknown inventor'
merged_df.loc[:, 'inventor_author'] = merged_df['inventor_author'].fillna('unknown inventor')

# grant_date -> fill with 'not granted yet' because the patent might still be pending
merged_df.loc[:, 'grant_date'] = merged_df['grant_date'].fillna('not granted yet')

# Numerical columns -> fill missing values with the mean (average) of the column
merged_df.loc[:, 'publication_year'] = merged_df['publication_year'].fillna(merged_df['publication_year'].mean())
merged_df.loc[:, 'filing_year'] = merged_df['filing_year'].fillna(merged_df['filing_year'].mean())
```

```python
# Calculate the total number of missing (null) values for each column in the dataframe
merged_df.isnull().sum()
```

```python
# Fill missing values in the 'image_link' column with the placeholder string 'no_image'
merged_df['image_link'] = merged_df['image_link'].fillna('no_image')
```

```python
# Calculate the total number of missing (null) values for each column in the dataframe
merged_df.isnull().sum()
```

```python
# Save the dataframe to a CSV file named 'patentAI.csv'
# index=False: Do not write row numbers to the file
# encoding='utf-8-sig': Ensures special characters (like Turkish, Chinese, etc.) are handled perfectly without errors
merged_df.to_csv('patentAI.csv', index=False, encoding='utf-8-sig')

# Print a confirmation message indicating the file has been saved
print("patentAI.csv kaydedildi.")
```
